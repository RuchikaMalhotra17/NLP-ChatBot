```python
# importing regex and random libraries
import re
import random
 
 
class AiBot:
    # potential negative responses
    negative_responses = ("no", "nope", "nah", "naw", "not a chance", "sorry")
    # keywords for exiting the conversation
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")
    # predefined questions
    predefined_questions = [
        "Where are you from? I'm always curious to learn about different places. ",
        "What's your favorite type of music? ",
        "What are some of your interests or hobbies? ",
        "What is your favourite cuisine? ",
        "Are we friends now? ",
        "Will we meet again?",
    ]
 
    def __init__(self):
        self.alienbabble = {'place': r'.*\s*you from*',
                            'music': r'.*\s*music*',
                            'hobbies': r'.*\s*like to do*',
                            'cuisine': r'.*\s*like to eat*'
                            }
        self.current_question_index = 0
 
    def greet(self):
        self.name = input("What is your name?\n")
        will_help = input(
            f"Hi {self.name}, I'm ChatBot. Will you help me learn about yourself?\n")
        if will_help in self.negative_responses:
            print("Ok, have a nice day! Nice talking to you.")
            return
        self.chat()
 
    def make_exit(self, reply):
        for command in self.exit_commands:
            if reply == command:
                print("Okay, have a nice day! Nice talking to you.")
                return True
 
    def chat(self):
        print(f"Nice to meet you, {self.name}!")
        # Present questions in a predefined sequence
        for question in self.predefined_questions:
            print(question)
            reply = input()
            if self.make_exit(reply):
                break
            self.respond_to_reply(reply)
        print("Bye! Good Luck")
 
    def respond_to_reply(self, reply):
        found_match = False
        for key, value in self.alienbabble.items():
            intent = key
            regex_pattern = value
            found_match = re.match(regex_pattern, reply)
            if found_match and intent == 'place':
                print(self.place())
                break
            elif found_match and intent == 'music':
                print(self.music())
                break
            elif found_match and intent == 'hobbies':
                print(self.hobbies())
                break
            elif found_match and intent == 'cuisine':
                print(self.cuisine())
                break
        if not found_match:
            print(self.no_match_intent())
 
    def place(self):
        responses = ("I am from the States.\n",
                     "I am from India.\n")
        return random.choice(responses)
 
    def music(self):
        responses = ("I like the blues.\n", "Pop music is the best.\n",
                     "I feel rock music is good.\n")
        return random.choice(responses)
 
    def hobbies(self):
        responses = ("I like reading.\n", "Golf!\n",
                     "I play Chess.\n")
        return random.choice(responses)
 
    def cuisine(self):
        responses = ("I like Indian.\n", "I like Italian.\n",
                     "I play Chinese.\n", "I like Mexican.\n")
        return random.choice(responses)
 
    def no_match_intent(self):
        responses = (
        "Thats nice to hear.\n", "Great to hear!\n", "I see.\n",
        "Interesting.")
        return random.choice(responses)
 
 
AiBot = AiBot()
AiBot.greet()
```
